```python
mp.title("stress-strain")
h=mp.plot(x,color="blue")
h=mp.plot(y,color="green")
print(h)
mp.xlabel('stress',color="blue")
mp.ylabel('strain',color="green")
mp.legend("ic")
mp.show()
```

```python
mp.title("stress-strain")
mp.bar(x,y,color=['red','purple','orange'])
h=mp.plot(x,color="blue",linestyle="dashed")
h=mp.plot(y,color="green",linestyle="dotted")
print(h)
mp.xlabel('stress',color="blue")
mp.ylabel('strain',color="green")
mp.legend("ic")
mp.show()
mp.xlim(1,60)
mp.ylim(1,60)
```
